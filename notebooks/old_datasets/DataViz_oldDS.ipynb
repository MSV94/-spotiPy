```python
Nombre_avis=df3[['track_id','hashtag','created_at']].groupby('track_id').count()
Nombre_avis
```

```python
Nombre_avis= Nombre_avis.rename(columns={'hashtag':"Nombre avis"})
Nombre_avis= Nombre_avis.rename(columns={'created_at':"Nombre de musique"})
```

```python
Nombre_avis2=Nombre_avis.groupby("Nombre avis").count()
Nombre_avis2['Nombre de hashtag']=Nombre_avis2.index
Nombre_avis2.head(20)
```

```python
plt.figure(figsize=(10,10))
sns.relplot(x='Nombre de hashtag',y='Nombre de musique',data=Nombre_avis2)
plt.yscale('log')
plt.savefig('NbM1.png')
plt.title('Nombre de musique en fonction du nombre de Hashtag qui leur sont associé')
plt.show();
```

```python
fig,axe = plt.subplots(1,2,figsize=(20,10))
axe[0].scatter(Nombre_avis2['Nombre de hashtag'].iloc[0:100],
                Nombre_avis2['Nombre de musique'].iloc[0:100],
                s=5,color='blue',marker='x')
axe[0].set_yscale('log')
axe[0].set_ylim([0,100000])
axe[0].set_xlim([0,100])
axe[0].set_xlabel('Nombres de hashtag')
axe[0].set_ylabel('Nombres de musique')
axe[0].set_title('Nombres de musique en fonction du nombres de Hashtag qui leur sont associé')

axe[1].scatter(Nombre_avis2['Nombre de hashtag'].iloc[100:-15],
               Nombre_avis2['Nombre de musique'].iloc[100:-15],
                s=1,color='blue',marker='x')

axe[1].set_yscale('log')
axe[1].set_xlim([0,16000])
axe[1].set_ylim([0,1000])
axe[1].set_xlabel('Nombre de hashtag')
axe[1].set_ylabel('Nombre de musique')

axe[1].set_title('Nombre de musique en fonction du nombre de Hashtag qui leur sont associé')
plt.savefig('NbM4.png')
plt.show();
```

```python
plt.figure()
sns.relplot(x='Nombre de hashtag',y='Nombre de musique',data=Nombre_avis2.iloc[0:100])
plt.yscale('log')
plt.savefig('NbM2.png')
plt.show();
```

```python
plt.figure()
sns.relplot(x='Nombre de hashtag',y='Nombre de musique',data=Nombre_avis2.iloc[100:-15])
plt.yscale('log')
plt.savefig('NbM3.png')
plt.show();
```

```python
functions_to_apply = {
    'hashtag' : ['nunique']}
Nombre_avis_unique=df3[['track_id','hashtag','created_at']].groupby('track_id').agg(functions_to_apply)
Nombre_avis_unique
```

```python
Nombre_avis_unique2=Nombre_avis_unique['hashtag'].value_counts()
```

```python
Nombre_avis_unique2=pd.DataFrame(Nombre_avis_unique2,columns=['Nombre de musique'])
```

```python
Nombre_avis_unique2.head()
```

```python
d1=np.array(Nombre_avis_unique2.index)
```

```python
d2=[]
for i,j in enumerate(d1):
    d2.append(d1[i][0])
```

```python
d2=np.array(d2)
```

```python
Nombre_avis_unique2['Nombre de hastag différents']=d2
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Nombre_avis_unique2['Nombre de hastag différents'],Nombre_avis_unique2['Nombre de musique'],
           s=3,color='blue',marker='x')
plt.yscale('log')
plt.ylim([0,200000])
plt.savefig('Nb_H_D.png')
plt.show();
```

```python
fig,axe = plt.subplots(1,2,figsize=(20,10))
axe[1].scatter(Nombre_avis_unique2['Nombre de hastag différents'].iloc[40:],
                Nombre_avis_unique2['Nombre de musique'].iloc[40:],
                s=3,color='blue',marker='x')
axe[1].set_ylim([0,100])
axe[1].set_xlim([0,600])
axe[1].set_xlabel('Nombre de hastag different')
axe[1].set_ylabel('Nombre de musique')
axe[1].set_title('Nombre de musique en fonction du nombre de Hashtag different qui leur sont associé')

axe[0].scatter(Nombre_avis_unique2['Nombre de hastag différents'],Nombre_avis_unique2['Nombre de musique'],
                s=3,color='blue',marker='x')

axe[0].set_yscale('log')
axe[0].set_xlim([0,600])
axe[0].set_ylim([0,200000])
axe[0].set_xlabel('Nombre de hastag different')
axe[0].set_ylabel('Nombre de musique')

axe[0].set_title('Nombre de musique en fonction du nombre de Hashtag different qui leur sont associé')
plt.savefig('Nb_H_D.png')
plt.show();
```
