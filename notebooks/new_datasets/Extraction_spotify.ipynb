```python
df = pd.read_csv('Data/tracks_features.csv')
df.head()
```

```python
# Fonction utilisé ensuite pour supprimer les accolade dans les lignes de "artist_ids"
def reduc(col):
    return ast.literal_eval(col)[0]
```

```python
# Récupere les genre musicaux et transforme la liste les contenant en chaine de caractère
def add_genre(col):
    artist_g=sp.artist(col)['genres']
    if artist_g==[]:
        return np.NaN
    else:
        return ' '.join([d+',' for d in artist_g])
```

```python
# Fonction récupérant la note de popularité d'une musique
def add_popularity(col):
    return sp.track(col)['popularity']
```
